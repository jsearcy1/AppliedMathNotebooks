```python
# this is the Code environemnt
# Here, everything after '#' is considered a comment
# and is not ignored by the computer.
# This opton allows for someone to comment their code.

# Remember, that it is important to comment your code.
# On the same level of importance as eating your vegetables!
```

```python
# to start a python notebook (the language in which you will be coding)
# you first import your libraries
# common libraries are

# numpy             - abreviated as np  (used in math)
# tensorflow        - abreviated as tf  (used in Machine Learning)
# pandas            - abreviated as pd  (used for table and database manipulations)
# matplotlib.pyplot - abreviated as plt ( used for plotting)

# to import a package you say: import [name of package]

import numpy as np # importing numpy as np for mathematical operations
import matplotlib.pyplot as plt # importing pyplot as plt for plotting
```

```python
# Once you imported your packages you should define your functions, if you're using some.
# Functions are great for keeping your code clean and avoid the situations of having 
# to write long list of repeating operations multiple times

def myFunction (var0, var1, var2):
    '''
    Everything here is also a comment.
    This is the function called my Function.
    It takes in 
        var0, var1, var2
    as variables for further computations.
    '''
    
    # Now I will tell my function what to do.
    # It will store the result of the instructions in a temporary
    # variable called result
    
    result = np.exp( -var1 * ( var0 + var2 )**2 )
    
    #now we return the value od the result  
    return result
```

```python
# What I want to do is find the value of the function 
# for a large range of values for x
# and fixed values for a = 0.5 and b = 3.0

a = 0.5 # declaring the value of a
b = 3.0 # declaring the variable b

# and then, fixed values for c = 0.1 and d = 2.0

c = 0.1
d = 2.0

# Now, I specify x as an numpy array.
# that is fancy lingo for a list of numbers
# I want a list of numbers from -20 to 20 in steps of 0.1
# to do that I use the arange function from the numpy library 
# (we imported it in the first cell)

x = np.arange(-10, 10, 0.1) 

# Now I compute the values of my function for 
# the specied range of x values and the fixed constants a, b then c, d
# we store that in a new variable called y0 and y1 respectively

y0 = myFunction(var0 = x , var1 = a, var2 = b) # Declaring new variable y0 as the results of my Function
y1 = myFunction(x , c, d) # Declaring new variable y1 as the results of my Function


#printing the first 10 values of y0
print('The first 10 values of y0 are:')
print(y0[0 : 10])
```

```python
# Now we will plot the function we just computed
# For this we use the pyplot = plt library

# declaring the start of my figure

plt.figure( figsize = (10,6) ) # figsize speicifes the (length , height) of figure

# now we plot the the results
# first the values of x and y0 and label it f(x, 0.5, 3.0)
plt.plot(x, y0, color = 'navy', linewidth = 2, label = f'var1 = {a} and var2 = {b}')
# first the values of x and y01and label it f(x, 0.1, 2.0)
plt.plot(x, y1, color = 'maroon', linewidth = 2, label = f'var1 = {c} and var2 = {d}')

# display my legend
plt.legend( fontsize = 10)

#Don't Forget to label your axis
plt.xlabel('x Values (no units)', fontsize = 12)        # x-axis label
plt.ylabel('y = f(x) Values (no units)', fontsize = 12) # y-axis label

# Title
plt.title(' My Gaussian Functions', fontsize = 15)

# Show my figure
plt.show()
```
